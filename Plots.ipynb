# Plotting

```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
#plot histogram
import matplotlib.pyplot as plt
tips.tip.plot(kind = 'hist')
```

```python
import matplotlib.pyplot as plt
tips.tip.plot(kind = 'hist')
plt.show
```

```python
#plot bar chart smoker

cts = tips.smoker.value_counts()
cts
```

```python
cts.plot(kind = 'bar')
```

```python
# count plot bar

sns.countplot(x = 'smoker' , data = tips)
```

```python
# distribution   for continuous variable columns

sns.distplot(tips.total_bill)
```

```python
# line graphs   #scatter plot

sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'smoker')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, row = 'day')
```

```python
# facet 
facet = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex')
```

```python
#
facet = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
